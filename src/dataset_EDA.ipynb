```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_url = '../data/android_data_latest'
subset = data_url + '/wlk/wlk1-SensorData.csv'
df = pd.read_csv(subset)
df = df.drop('Unnamed: 0', axis=1) if 'Unnamed: 0' in df.columns else df
df = df.drop('id', axis=1) if 'id' in df.columns else df
df
```

```python
# check for duplicates
duplicates = df.duplicated()
print('Number of duplicates: ', duplicates.sum())
```

```python
# remove duplicates
df = df.drop_duplicates()
df
```

```python
# check for missing values
missing = df.isnull().sum()
print('Number of missing values: ', missing.sum())
```

```python
# check duplicates
duplicates = df.duplicated()
print('Number of duplicates: ', duplicates.sum())
```

```python
# save data to csv with same name and directoty
df.to_csv(subset, index=False)
```

```python
plt.boxplot(df)
plt.show()
```

```python
def remove_outliers(df):
    result = df.copy()
    for feature in df.columns:
        if pd.api.types.is_numeric_dtype(df[feature]):
            Q1 = df[feature].quantile(0.25)
            Q3 = df[feature].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            result = result[(result[feature] >= lower_bound) & (result[feature] <= upper_bound)]
    return result
```

```python
df = remove_outliers(df)
```
